```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot.
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create a plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#create labels for x and y axis.
plt.xlabel('Date')
plt.ylabel('Fare($)')

#Set the y limit between 0 and 45
plt.ylim(0,45)

#create a title
plt.title('PyBer Fare by Month')

#Add agrid.
plt.grid()

#Add the legend
plt.legend()
```

```python

fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,color = 'skyblue', s = [item * 5 for item in y_axis], edgecolor = 'black', linewidth = 2, label = 'Boston', alpha = 0.2)


#Change the size of the markers to 5 times each data point.


#Make the color 20% transparent.


ax.invert_yaxis()
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)

```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "indianred", "greenyellow", "orange", "gold", "tomato", "yellow", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "indianred", "greenyellow", "orange", "gold", "tomato", "yellow", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
    explode=explode_values,
    startangle=90,
    counterclock = False,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()

```
